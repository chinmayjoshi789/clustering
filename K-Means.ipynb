```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("D:/student_evaluation.csv")
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df)
```

```python
pred=kmeans.predict(df)
```

```python
pred
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

```python
SSE = []
```

```python
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(df)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.dfFrame({'Cluster':range(1,20), 'SSE':SSE})
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = scaler.fit_transform(df)
```

```python
pd.dfFrame(df_scaled).describe()
```

```python
SSE_scaled = []
```

```python
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(df_scaled)
    SSE_scaled.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12,6))
frame_scaled = pd.dfFrame({'Cluster':range(1,20), 'SSE':SSE_scaled})
plt.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker='o')
plt.xlabel("Clusters")
plt.ylabel("SSE")
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 4)
kmeans.fit(df_scaled)
pred = kmeans.predict(df_scaled)
```
